***Functions***
A function is a device that groups a set of statements so they can be run more than once in a programâ€”a packaged procedure invoked by name.





1.   Two ways to make functions (def and lambda),
 
2.   Two ways to manage scope visibility (global and nonlocal)

3.   Two ways to
send results back to callers (return and yield)












***Why Use Functions?***


1.   Maximizing code reuse and minimizing redundancy
2.   Procedural decomposition






***Coding Functions***-Brief introduction to the main concepts behind Python functions


1.   def is executable code.
2.   def creates an object and assigns it to a name.
3.   lambda creates an object but returns it as a result.
4.   return sends a result object back to the caller.
5.   yield sends a result object back to the caller, but remembers where it left
off.
6.   global declares module-level variables that are to be assigned.
7.   nonlocal declares enclosing function variables that are to be assigned.
8.   Arguments are passed by assignment (object reference).
7.   Arguments are passed by position, unless you say otherwise.
8.   Arguments, return values, and variables are not declared.









***Advantages of functions***


1.   Putting the code in a function makes it a tool that you can run as many times as
you like.

2.   Because callers can pass in arbitrary arguments, functions are general enough to
work on any two sequences (or other iterables) you wish to intersect.


3.   When the logic is packaged in a function, you have to change code in only one
place if you ever need to change the way the intersection works.

4.   Coding the function in a module file means it can be imported and reused by any
program run on your machine.

```python
times(2, 4) # Arguments in parentheses
```

```python
x = times(3.14, 4) # Save the result object
x
```

```python
def intersect(seq1, seq2):
  res = [] # Start empty
  for x in seq1: # Scan seq1
   if x in seq2: # Common item?
     res.append(x) # Add to end
  return res
```

```python
s1 = "SPAM"
s2 = "SCAM"
intersect(s1, s2) # Strings
```

```python
x = intersect([1, 2, 3], (1, 4)) # Mixed types
x # Saved result object
```

```python
def vowel_count(s):
    VOWELS = "aeiouAEIOU"
    vc = 0
    for ch in s:
        if ch in VOWELS:
            vc += 1
    return vc
print(vowel_count("Python is the best platform to learn programming language."))
```
