lists can contain any sort of object:
numbers, strings, and even other lists. 
unlike strings, lists may be changed in place by assignment to offsets and slices, list method calls, deletion statements, they are mutable objects.

Python lists are:
Ordered collections of arbitrary objects
Accessed by offset
Variable-length, heterogeneous, and arbitrarily nestable
Of the category “mutable sequence”
Arrays of object references

List Iteration and Comprehensions

```python
4 in [1, 2, 3]
```

```python
for x in [1, 2, 3]:
  print(x, end='')
```

```python
res = [c * 4 for c in 'SPAMx']
res
```

```python
res = []
for c in 'SPAM':                # List comprehension equivalent
    res.append(c * 4)
res
```

the map built-in function does similar work, but applies a function to items in a sequence and collects all the results in a new list:

```python
list(map(abs, [-1, 2, 0, 1, 2]))       # Map a function across a sequence
```
